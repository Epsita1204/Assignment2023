```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(-5,5,100)
y=2*x+1
plt.plot(x,y,'-b',label='y=2x+1')
plt.title('Graph of y=2x+1')
plt.xlabel('x',color='g')
plt.ylabel('y',color='r')
plt.legend(loc='upper left')
plt.grid()
display(plt.show())
```
